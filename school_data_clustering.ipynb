# <h1>Data Preprocessing</h1>


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('school_data.csv')
df.tail()
```

Select sekolah yang diakreditasi tahun 2010 keatas, dan memiliki peringkat akreditasi

```python
nullIndexes = df.loc[(df['Tahun Akreditasi']<2010) | (df['Peringkat Akreditasi'].isnull()) | (df['Nilai Akreditasi']==0)].index.values.tolist()
```

```python
df_nonull = df.drop(nullIndexes)
df_nonull = df_nonull.reset_index()
df_nonull
```

Drop data tidak penting ('Tahun Akreditasi', 'Last Sync', 'Jml Sync') dan juga SLB (Sekolah Luar Biasa) karena dapat mengganggu clustering

```python
df_nonull = df_nonull.drop(columns = ['Tahun Akreditasi', 'Last Sync', 'Jml Sync', 'index'])
df_nonull = df_nonull.drop(df_nonull.loc[df_nonull['BP']=='SLB'].index.values.tolist())
```

Ubah data categorical menjadi numerik (kota, kecamatan, BP, status, peringkat akreditasi)

```python
from sklearn.preprocessing import LabelEncoder

def cat_to_num(col_name):
  enc = LabelEncoder()
  enc.fit(df_nonull[col_name])

  label_values = df_nonull[col_name].unique()

  df_nonull[col_name] = enc.fit_transform(df_nonull[col_name])
  df_nonull[col_name] = df_nonull[col_name].astype(int)

  key_values = df_nonull[col_name].unique()
  print(len(key_values))
  dict_label = {}
  for i in range(len(key_values)):
    dict_label[key_values[i]] = label_values[i]

  return dict_label
```

```python
categorical_col = ['Kota/Kabupaten', 'Kecamatan', 'BP', 'Status', 'Peringkat Akreditasi']
dict_categories = [cat_to_num(categorical_col[i]) for i in range (len(categorical_col))]
```

```python
dict_categories
```

```python
df_nonull.head()
```

```python
df = df_nonull
df.describe()
```

Drop data non numerik

```python
df1 = df.drop(columns = ['Nama Sekolah', 'NPSN', 'Kota/Kabupaten', 'Kecamatan', 'BP', 'Status', 'Peringkat Akreditasi'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df1)
```

```python
df1 = pd.DataFrame(scaled)
```

```python
df1.head()
```

# <h1>K-Means Clustering</h1>

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## Cari nilai k terbaik

Elbow method untuk mencari nilai k terbaik

```python
distortions = []
K = range(1,10)
for k in K:
  kmeanModel = KMeans(n_clusters=k)
  kmeanModel.fit(df1)
  distortions.append(kmeanModel.inertia_)

plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method to get optimal k')
plt.show()
```

Grid Search untuk mencari nilai k terbaik

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
param_grid = {'n_clusters': [2, 3, 4, 5, 6, 7, 8, 9, 10]}
scoring_metric = silhouette_score

kmeans = KMeans()
grid_search = GridSearchCV(kmeans, param_grid=param_grid, scoring=scoring_metric)

grid_search.fit(df1)

print("Best hyperparameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

# DBSCAN

```python
df1.head()
```

```python
# Minimum samples (“MinPts”): the fewest number of points required to form a cluster
# ε (epsilon or “eps”): the maximum distance two points can be from one another while still belonging to the same cluster
# min_sample terbaik biasanya minimal 2*jumlah feature, sehingga sekitar 16
# plotting k-distance graph untuk mencari nilai eps optimal
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt

neighbors = NearestNeighbors(n_neighbors=16)
neighbors_fit = neighbors.fit(df1)
distances, indices = neighbors_fit.kneighbors(df1)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.xlim(3900,4200)
plt.axhline(y = 0.32, color='r')
plt.plot(distances)
```

##Clustering

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=0.175, min_samples=1000, metric='euclidean').fit(df1)
cluster_dbscan = clustering.labels_
print(np.count_nonzero(cluster_dbscan == -1))
print(np.count_nonzero(cluster_dbscan == 0))
np.unique(cluster_dbscan)
```

```python
print(cluster_dbscan)
```

```python
df['db_cluster'] = cluster_dbscan
df.tail()
```

```python
plt.pie(df.groupby('db_cluster').size(), labels =['outliers','cluster 0'], autopct='%1.1f%%')
plt.title('Number of school per cluster')
```

```python
data = df1.to_numpy()
clusters = df['db_cluster'].to_numpy()
s = silhouette_score(data, clusters, metric="euclidean")
db = davies_bouldin_score(data, clusters)
```
