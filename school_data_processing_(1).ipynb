This notebook collects individual files from the government's website and compiles it to 1 csv (originally ran on google colab)

```python
import pandas as pd
```

csv is compiled from the excel files downloaded from https://dapo.kemdikbud.go.id/sp/1/010000

```python
df = pd.read_csv('/content/mess_barat.csv', sep=';', encoding = 'unicode_escape')
df.head()
```

```python
df.columns = ['Kecamatan', 'Nama Sekolah', 'NPSN', 'BP', 'Status', 'Last Sync', 'Jml Sync', 'PD', 'Rombel', 'Guru', 'Pegawai', 'R. Kelas', 'R. Lab', 'R. Perpus']
df.head()
```

```python
df.drop(df[df['Nama Sekolah'] == 'Nama Sekolah'].index, inplace=True)
df.drop(df[df['Nama Sekolah'] == 'Total'].index, inplace=True)
df.head()
```

```python
df['Kecamatan'] = df['Kecamatan'].str[24:-21]
df.head()
```

```python
kota = 'Jakarta Barat'
df.insert(0, 'Kota/Kabupaten', kota)
df.head()
```

```python
df.dtypes
```

```python
df[['Jml Sync', 'PD', 'Rombel', 'Guru', 'Pegawai', 'R. Kelas', 'R. Lab', 'R. Perpus']] = df[['Jml Sync', 'PD', 'Rombel', 'Guru', 'Pegawai', 'R. Kelas', 'R. Lab', 'R. Perpus']].astype(int)
df.dtypes
```

```python
df.describe()
```

```python
filename=kota + '_processed.csv'
df.to_csv(filename, index=False)
print('done')
```

```python
df2 = pd.read_csv('/content/merged.csv')
df2.describe()
```

```python
df2.drop(df2[df2['Kota/Kabupaten'] == 'Kota/Kabupaten'].index, inplace=True)
df2.describe()
```

```python
df2.dtypes
```

```python
df2[['Jml Sync', 'PD', 'Rombel', 'Guru', 'Pegawai', 'R. Kelas', 'R. Lab', 'R. Perpus']] = df2[['Jml Sync', 'PD', 'Rombel', 'Guru', 'Pegawai', 'R. Kelas', 'R. Lab', 'R. Perpus']].astype(int)
df2['Last Sync'] = pd.to_datetime(df2['Last Sync'], format='%d %b %Y %H:%M:%S', errors='coerce')
```
