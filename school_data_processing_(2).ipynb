This notebook collects accreditation data from https://bansm.kemdikbud.go.id/ and adds it to the existing dataset from first notebook (Originally ran on google colab)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/school_data.csv')
# df.insert(6, "Tahun Akreditasi", 0)
# df.insert(7, "Nilai Akreditasi", 0)
# df.insert(8, "Peringkat Akreditasi", np.nan)
# df[['Tahun Akreditasi', 'Nilai Akreditasi']] = df[['Tahun Akreditasi', 'Nilai Akreditasi']].astype(int)
# df['Peringkat Akreditasi'] = df['Peringkat Akreditasi'] .astype(object)
# df.dtypes
```

```python
import requests
from bs4 import BeautifulSoup
import re

for i in range(2685, len(df)):
# for x in range(1):
  # i = 369
  school_id=df.loc[i, 'NPSN']
  URL = "https://bansm.kemdikbud.go.id/home/sekolah/"+str(school_id)
  page = requests.get(URL)
  print(URL)
  soup = BeautifulSoup(page.content, "html.parser")
  result = soup.findAll('div', 'col-xs-12 col-md-4 text-left')
  if len(result)!=0:
    data=result[-1]
    # tahun akreditasi
    tahun=data.findAll('b')
    tahun_akred=0
    for j in tahun:
      x=re.findall(r'\d+', j.text)
      if(len(x)!=0):
        tahun_akred=x
        break

    # nilai akreditasi
    nilai=data.findAll("font", {"class": "text-info"})
    nilai_akred=nilai[-2]

    # akreditasi
    akreditasi=nilai[-1]

    df.loc[i, 'Peringkat Akreditasi']=akreditasi.text
    if(akreditasi.text != 'TT'):
      df.loc[i, 'Tahun Akreditasi']=int(tahun_akred[0])
      df.loc[i, 'Nilai Akreditasi']=int(nilai_akred.text)

    print('added')
```
